```python
from transformers import AutoTokenizer
```

```python
base_model = "meta-llama/Llama-2-7b-hf"
```

```python
from huggingface_hub import login
login()
```

```python
tokenizer = AutoTokenizer.from_pretrained(
    base_model,
    padding_side = "right",
    add_eos_token = True,
)
tokenizer.pad_token = tokenizer.eos_token
tokenizer.add_bos_token, tokenizer.add_eos_token
```

```python
tokenizer
```

```python
import torch
from transformers import BitsAndBytesConfig

bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_use_double_quant=False,
    bnb_4bit_compute_dtype=torch.bfloat16
)
```

```python
from transformers import AutoModelForCausalLM

model = AutoModelForCausalLM.from_pretrained(
    base_model,
    load_in_4bit=True,
    quantization_config=bnb_config,
    torch_dtype=torch.bfloat16,
    device_map="auto",
)
```

```python
model
```

```python
from datasets import load_dataset

dataset_name = "sciq"
```

```python
train_dataset = load_dataset(dataset_name, split = "train[0:1100]")
eval_dataset = load_dataset(dataset_name, split="validation[0:100]")
```

```python
train_dataset.to_pandas().head(10)
```

```python
train_dataset.to_pandas().dtypes
```

```python
def generate_prompt(sample):
    full_prompt =f"""<s>[INST]{sample['question']}
    {f"Here is some info: {sample['support']}" if len(sample["support"]) > 0 else None}
    [/INST] {sample['correct_answer']}</s>"""
    return {"text": full_prompt}
```

```python
generate_prompt(train_dataset[0])
```

```python
generated_train_dataset = train_dataset.map(
    generate_prompt, remove_columns=list(train_dataset.features))
generated_val_dataset = eval_dataset.map(
    generate_prompt, remove_columns=list(eval_dataset.features))
```

```python
generated_train_dataset[5]["text"]
```

```python
tokenizer(generated_train_dataset[5]["text"])
```

```python
from peft import prepare_model_for_kbit_training

model.gradient_checkpointing_enable()

model = prepare_model_for_kbit_training(model)
```

```python
def print_trainable_parameters(model):
    trainable_params = 0
    all_param = 0
    for _, param in model.named_parameters():
        all_param += param.numel()
        if param.requires_grad:
            trainable_params += param.numel()
    print(
        f"trainable params: {trainable_params} || all params: {all_param} || trainable%: {100 * trainable_params / all_param}"
    )
```

```python
from peft import LoraConfig, get_peft_model

lora_config = LoraConfig(
    r=8,
    lora_alpha=16,
    target_modules=[
        "q_proj",
        "k_proj",
        "v_proj",
        "o_proj",
        "gate_proj",
        "up_proj",
        "down_proj",
        "lm_head",
    ],
    bias="none",
    lora_dropout=0.05,
    task_type="CAUSAL_LM",
)
```

```python
from peft import get_peft_model

model = get_peft_model(model, lora_config)

print_trainable_parameters(model)
```

```python
model
```

```python
from transformers import TrainingArguments

training_arguments = TrainingArguments(
    output_dir="./results",
    num_train_epochs=2,
    per_device_train_batch_size=4,
    gradient_accumulation_steps=1,
    optim="paged_adamw_32bit",
    save_strategy="steps",
    save_steps=25,
    logging_steps=25,
    learning_rate=2e-4,
    weight_decay=0.001,
    max_steps=50,
    evaluation_strategy="steps",
    eval_steps=25,
    do_eval=True,
    report_to="none",
)
```

```python
from trl import SFTTrainer

# Setting sft parameters
trainer = SFTTrainer(
    model=model,
    tokenizer=tokenizer,
    args=training_arguments,
    train_dataset=generated_train_dataset,
    eval_dataset=generated_val_dataset,
    peft_config=lora_config,
    dataset_text_field="text",
)
```

```python
model.config.use_cache = False
trainer.train()
```

```python
my_finetuned_model = "Llama-2-7b-hf-SciQ"

trainer.model.push_to_hub(my_finetuned_model)
```
